# Packages

```python
import numpy as np
from scipy.special import i0, iv
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

# von Mises distribution

```python
p=2                                            # concentration parameter von Mises
q=np.linspace(-0.5,1.5,500)                    # x/L coordinate along the ring
vmises=1/i0(p)*np.exp(p*np.cos(2*np.pi*q))     # von Mises distribution
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(q,vmises, color = 'deepskyblue')
ax.set_xlabel(r"$x/L$", fontsize = 30)
ax.set_ylabel(r"$\frac{G(x)}{g}$", fontsize = 40, rotation = 0, labelpad = 50)
ax.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)
plt.tight_layout()
ax.spines[['right', 'top']].set_visible(False)
ax.tick_params(axis='both', which='major', labelsize=15)
for axis in 'left', 'bottom':
    ax.spines[axis].set_linewidth(1.5)
plt.show()
```

# Friction coefficient in Fourier space

```python
def nun(N,x,n,y,p):                                               # nu_n/nu_A at different y = c/v_0
    return N*(x**2/(np.pi**2*n**2*y**2)-1)*y**2*iv(n,p)/i0(p)
def passive(N,x,n,p):                                             # passive limit of nu_n/nu_A
    return N*x**2/(n**2*np.pi**2)*iv(n,p)/i0(p)

x=np.linspace(0,8,500)                                            # x = L alpha c/v_0^2
n=1                                                               # Fourier mode number
N=10                                                              # number of RTP colloids
p=2                                                               # concentration parameter von Mises
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x,nun(N,x,n,1,p), label = r"$\frac{c}{v_0}=1$", color = 'black')
ax.plot(x,nun(N,x,n,0.75,p), label = r"$\frac{c}{v_0}=0.75$", color = 'lime')
ax.plot(x,nun(N,x,n,0.5,p), label = r"$\frac{c}{v_0}=0.5$", color = 'orange')
ax.plot(x,nun(N,x,n,0.3,p), label = r"$\frac{c}{v_0}=0.75$", color = 'darkviolet')
ax.plot(x,passive(N,x,n,p), label = "Passive limit", color = 'cyan', linestyle = '--')
ax.set_xlabel(r"$L \alpha c/v_0^2$", fontsize = 30)
ax.set_ylabel(r"$\frac{\nu_1}{\nu_A}$", fontsize = 40, rotation = 0, labelpad = 40)
ax.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)
plt.tight_layout()
ax.set_ylim([-8, 30])
ax.legend(fontsize = 15)
ax.spines[['right', 'top']].set_visible(False)
ax.tick_params(axis='both', which='major', labelsize=15)
for axis in 'left', 'bottom':
    ax.spines[axis].set_linewidth(1.5)
plt.show()
```
